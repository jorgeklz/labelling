FUNCIONES

```
#Convierto cada tweet del dataset en un vector de palabras
from nltk import word_tokenize
def ConvertirData(corpusBase):
  textos = []
  for y in corpusBase['text']:
    doc = word_tokenize(y)
    textos.append(doc)
  return textos
```

RECURSOS

```
Data_Final = pd.read_excel('/content/gdrive/My Drive/Colab Notebooks/Reporte_tweets_Ecuador/4DataFinalLemaFiltro.xlsx')
```

**MODELOS**

MODELO COSENO

```
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```
TextData = list(Data_Final['text']) #Establesco todo los tweets en formato de lista
#TextData = lista_tweet()
vectorizer = CountVectorizer() 
texto = vectorizer.fit_transform(TextData) #Se hace el conteo de palabras mediante la data
```

```
#GUARDAR MODELO COSENO

#Damos el nombre al fichero
filename = '/content/gdrive/My Drive/Colab Notebooks/OTROSMODELOS/N_VectorizerCOSENO.pickle'

#Guardamos el vector
pickle.dump(vectorizer, open(filename, 'wb'))

#Damos el nombre al fichero
filename = '/content/gdrive/My Drive/Colab Notebooks/OTROSMODELOS/N_DataVectoCOSENO.pickle'

#Guardamos el data vectorizada
pickle.dump(texto, open(filename, 'wb'))
```

MODELO EMPLEANDO TFIDF

```
from gensim import corpora
from gensim.models import TfidfModel
from gensim.similarities import SparseMatrixSimilarity
```

```
texts = ConvertirData(Data_Final) #Se tokeniza cada tweet de la data
dictionary = corpora.Dictionary(texts) #Se genera el diccionario en base a la data
corpus = [dictionary.doc2bow(text) for text in texts]
tfidf = TfidfModel(corpus) #Se crea el modelo de frecuencia de palabras  
feature_cnt = len(dictionary.token2id) #Se obtiene la longitud del diccionario
index = SparseMatrixSimilarity(tfidf[corpus], num_features = feature_cnt) #Se crea la matriz de similaridad 
```

```
#GUARDAR MODELO TFIDF

#MODELO
#Damos el nombre al fichero
filename = '/content/gdrive/My Drive/Colab Notebooks/OTROSMODELOS/N_modelTFIDF.model'

#Guardamos el modelo
pickle.dump(tfidf, open(filename, 'wb'))


#DICCIONARIO
#Damos el nombre al fichero
filename = '/content/gdrive/My Drive/Colab Notebooks/OTROSMODELOS/N_DiccionarioTFIDF.pickle'

#Guardamos el diccionario
pickle.dump(dictionary, open(filename, 'wb'))


#MATRIZ DE SIMILARIDAD
#Damos el nombre al fichero
filename = '/content/gdrive/My Drive/Colab Notebooks/OTROSMODELOS/N_MatrizSimilaridadTFIDF.pickle'

#Guardamos la matriz de similaridad
pickle.dump(index, open(filename, 'wb'))
```
